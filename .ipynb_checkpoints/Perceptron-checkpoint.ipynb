```python
import numpy as np 
import keras
from keras.models import Sequential #this is to import the sequential model
from keras.layers import Dense #this adds the dense layers of network
from tensorflow.keras.optimizers import Adam #adam helps to learn the deep learning model
import matplotlib.pyplot as plt #to plot the graphs
%matplotlib inline
```

```python
#making the datasets of five hunred points 
n_pts = 500
np.random.seed(0)
Xa = np.array([np.random.normal(13, 2, n_pts),
               np.random.normal(12, 2, n_pts)]).T
Xb = np.array([np.random.normal(8, 2, n_pts),
               np.random.normal(6, 2, n_pts)]).T
#taking transpose to make the matrix eligible for the matrix calculations or to make the dimensions equal
X = np.vstack((Xa, Xb))
y = np.matrix(np.append(np.zeros(n_pts), np.ones(n_pts))).T
#plotting the points
plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
```

```python
#making the deep learning model 
model = Sequential()
model.add(Dense(units = 1, input_shape=(2,), activation = 'sigmoid')) #unit shows number of layers input shape number of nodes
adam = Adam(learning_rate = 0.1)
model.compile(adam,loss='binary_crossentropy',metrics=['accuracy']) #using the cross entropy to minimize the error
h = model.fit(x=X,y=y, verbose=1,batch_size=50, epochs=500,shuffle='true') #compiling the model
```

```python
#showing the accuracy as the number of epochs increases
# Epoch refers to one complete pass of the training dataset through the algorithm.
plt.plot(h.history['accuracy'])
plt.title('accuracy')
plt.xlabel('epoch')
plt.legend(['accuracy'])
```

```python
plt.plot(h.history['loss'])
plt.title('loss')
plt.xlabel('epoch')
plt.legend(['loss'])
```

```python
def plot_decision_boundary(X,y,model):
    #x_span and y_span are simply the x and y coordinates value on the graph the maximum value and 50 is number of subdivison
    x_span = np.linspace(min(X[:,0])-1,max(X[:,0])+1,50)
    y_span = np.linspace(min(X[:,1])-1,max(X[:,1])+1,50)
    xx, yy = np.meshgrid(x_span,y_span)
    xx_, yy_ = xx.ravel(), yy.ravel() #flattening our data for better computation
    grid = np.c_[xx_,yy_] 
    pred_func = model.predict(grid) #passing the data into our model
    z = pred_func.reshape(xx.shape) # Unflattening our data to be visually understandble
    plt.contourf(xx,yy,z)
    
    
```

```python
 #plotting the point again to show here
plot_decision_boundary(X,y,model)

plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
```

```python
plot_decision_boundary(X,y,model)

plt.scatter(X[:n_pts,0], X[:n_pts,1])
plt.scatter(X[n_pts:,0], X[n_pts:,1])
#defining a point to check whether or not the prediction is correct
x = 7.5
y = 5
point = np.array([[x,y]])
prediction = model.predict(point)
plt.plot([x],[y],marker = "o", markersize=10, color="red")
print("prediction is :", prediction)
```
